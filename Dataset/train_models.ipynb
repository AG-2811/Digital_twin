```python
import pandas as pd
import joblib
import tensorflow as tf
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Load datasets
activity = pd.read_csv("dailyActivity_merged.csv")
sleep = pd.read_csv("sleepDay_merged.csv")

# Convert date
activity["ActivityDate"] = pd.to_datetime(activity["ActivityDate"])
sleep["SleepDay"] = pd.to_datetime(sleep["SleepDay"])

# Merge
data = pd.merge(activity, sleep, on="Id", how="left")

# Fill missing sleep
data["TotalMinutesAsleep"].fillna(400, inplace=True)

# рџ”Ґ Add AGE (random/simulated for training)
import numpy as np
data["Age"] = np.random.randint(18, 60, size=len(data))

# Features
features = [
    "TotalSteps",
    "VeryActiveMinutes",
    "SedentaryMinutes",
    "Calories",
    "TotalMinutesAsleep",
    "Age"
]

X = data[features]

# рџЋЇ Mood Logic (label creation)
def mood_label(row):
    if row["VeryActiveMinutes"] < 10:
        return 0   # Sad
    elif row["VeryActiveMinutes"] < 40:
        return 1   # Stressed
    elif row["VeryActiveMinutes"] < 80:
        return 2   # Happy
    else:
        return 3   # Energetic

data["Mood"] = data.apply(mood_label, axis=1)

y = data["Mood"]

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# ================= ML MODEL =================
ml_model = RandomForestClassifier()
ml_model.fit(X_train, y_train)

joblib.dump(ml_model, "ml_model.pkl")
print("ML Model Saved")

# ================= DL MODEL =================
dl_model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation="relu", input_shape=(6,)),
    tf.keras.layers.Dense(32, activation="relu"),
    tf.keras.layers.Dense(16, activation="relu"),
    tf.keras.layers.Dense(4, activation="softmax")
])

dl_model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

dl_model.fit(X_train, y_train, epochs=20)

dl_model.save("dl_model.h5")
print("DL Model Saved")
```
